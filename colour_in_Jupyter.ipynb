```python
import matplotlib.pyplot as plt
import numpy as np

fig_w_inch = 18
fig_h_inch = 12
fig_dpi = 300

fig, axes = plt.subplots(figsize = [fig_w_inch, fig_h_inch], dpi = fig_dpiframeon = False)
fig.set_size_inches(fig_w_inch, fig_h_inch)
plt.axis('off')
plt.grid(b = None)
plt.margins(0)

axes.set_facecolor('#000000')
# axes.set_axis_bgcolor('#000000')
fig_w = fig_w_inch * fig_dpi
fig_h = fig_h_inch * fig_dpi

# pattern is dot, then 3 dots worth of space, then dot...
# first and last are half-dot
vis_dots_per_w = 16
vis_dots_per_h = vis_dots_per_w * fig_h_inch / fig_w_inch
invis_dots_between = 3
dots_per_w = (vis_dots_per_w - 1) * (invis_dots_between + 1) 
dot_d = fig_w / dots_per_w
dot_r = dot_d / 2
vis_dot_inter_center_spacing = dot_d * (invis_dots_between + 1)
vis_dot_origin = [ 0, 0 ]
vis_dot_max_x = vis_dot_origin[0] + vis_dot_inter_center_spacing * (vis_dots_per_w - 1)
vis_dot_max_y = vis_dot_origin[1] + vis_dot_inter_center_spacing * ((vis_dots_per_w * fig_h_inch / fig_w_inch) - 1)

dot_x_range = np.linspace(vis_dot_origin[0], vis_dot_max_x, vis_dots_per_w)
dot_y_range = np.linspace(vis_dot_origin[1], vis_dot_max_y, vis_dots_per_h)
dot_colors = [ 'r', 'g', 'b' ]

dot_x = []
dot_y = []
dot_color = []
for iy, y in enumerate(dot_y_range):
    color_cycle_offset = iy % len(dot_colors)
    for ix, x in enumerate(dot_x_range):
        dot_x.append(x)
        dot_y.append(y)
        dot_color.append(dot_colors[(color_cycle_offset + ix) % len(dot_colors)])

plt.scatter(dot_x, dot_y, dot_d, dot_color)
plt.savefig('/tmp/first_try.png', bbox_inches = 'tight', dpi = fig_dpi, facecolor = '#000000' )
```
